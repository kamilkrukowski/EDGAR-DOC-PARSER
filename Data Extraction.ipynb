### 3. News Catcher
Trial API (until 10/21): d1AJqQ0RVS1QmJebS64RV_y0ehdH-WaApkPPs2nVa1I <br/>
https://docs.newscatcherapi.com/api-docs/quick-start

```python


def Newscatcher_search(query, api_key):
    url = "https://api.newscatcherapi.com/v2/search"

    querystring = {"q":"\""+ query+ "\"","lang":"en","sort_by":"relevancy","page":"1"}

    headers = {
        "x-api-key": api_key
        }

    response = requests.request("GET", url, headers=headers, params=querystring)
    return response

```

```python
query = "IBM"
api_key = "d1AJqQ0RVS1QmJebS64RV_y0ehdH-WaApkPPs2nVa1I"
response = Newscatcher_search(query, api_key)
```
